# 1. Calcuate the Root Mean Square Error (RMSE)


**(a) Define the RMSE Function**  
 
Create a function that takes in two numerical arrays of the same length (think about one array
as actual values and the other as predicted values) and calculates the RMSE between them.

```python
# RMSE function
def calc_rmse(actual, predicted): 
    
    # Converted inputs into numpy arrays 
    actual = np.array(actual)
    predicted = np.array(predicted)

    # Squared_diff variable finds the difference and squares it
    squared_diffs = (predicted - actual) ** 2 

    # Grabs the average of the squared differences
    average_squared = np.mean(squared_diffs)

    # Square root the average 
    rmse = np.sqrt(average_squared)

    return rmse 
```

**(b) Apply RMSE function to sale_price & sale_price_pred**   

Apply your RMSE function to the columns “sale price” and “sale price pred” in the file hous-
ing data.csv. file and report your result. The sale price has been predicted from the square
footage of the house.

```python
# Loading the housing_data 
df_rmse = pd.read_csv('housing_data.csv')

# Grabs 'sale_price', and 'sale_price_pred' from the housing_data
actual_prices = df_rmse['sale_price']
predicted_prices = df_rmse['sale_price_pred']

# Applying the calc_rmse function to the two columns 
final_rmse = calc_rmse(actual_prices, predicted_prices)

# Printing the resutls 
print(f"The models Rmse is: ${final_rmse:,.2f}")
 
```

# 2. Calculate the Mean absolute Error (MAE)

**(a) Repeat problem (1a) above with MAE.**   
Create a function that takes in two numerical arrays of the same length (think about one array
as actual values and the other as predicted values) and calculates the MAE between them.


```python
# MAE function
def calc_mae(actual, predicted):

    # Converted inputs into numpy arrays
    actual = np.array(actual)
    predicted = np.array(predicted)

    # Subtract predicted from actual and get the absolute values
    abs_diffs = np.abs(predicted - actual)

    # Gets the average of the absolute values 
    mean_abs = np.mean(abs_diffs)

    return mean_abs
```

# 3. Percentage Accuracy 

**(a) Two binary array functions**  

Build a function that takes in two binary arrays of the same length, where one is actual values
and the other predicted values, that calculates the prediction accuracy.  


```python
# Accuracy function  
def accuracy(actual, predicted):

    # Converted inputs into numpy arrays 
    actual = np.array(actual)
    predicted = np.array(predicted)

    # Compare actual vs predicted
    match_accuracy = actual == predicted 
    
    # Cacluate the percentage of matches 
    percentage = np.sum(match_accuracy) / len(actual) * 100

    return percentage
```

**(b) Apply functions**  
Apply your accuracy function to the columns “actual” and “predicted” in the mushroom data.csv
file and report your result. The columns in this data set indicate the mushroom is poisonous with
a “p” and if it is edible with an “e”.     

Note: The mushroom data.csv data set is modified version of the agaricus-lepiota.data file from
https://archive.ics.uci.edu/ml/datasets/mushroom.

```python
# Loading the mushroom_data
df_mushroom = pd.read_csv('mushroom_data.csv')

# Applying the accuracy function to the two columns 
mushroom_accuracy = accuracy(df_mushroom['actual'], df_mushroom['predicted'])

# Printing the results 
print(f"Mushroom Accuracy: {mushroom_accuracy:.2f}%")
```

# 4. Creating a graph 

**(a) Matplotlib | (b) Value of p | (c) Minimum Error**

- (a) Use the Python package matplotlib to plot this function.
- (b) Use your graph to estimate the value of p that minimizes the error. Report your result.
- (c) Use your graph to estimate the minimum error. Report your result.


```python
# Importing libraries 
import matplotlib.pyplot as plt 
import numpy as np

# Create evenly spaced p-values from 0 to 10 
p = np.linspace(0, 10, 500)

# Calculate the error value f(p) for each point in p
y = 0.005*p**6 - 0.27*p**5 + 5.998*p**4 - 69.919*p**3 + 449.17*p**2 - 1499.7*p + 2028

# Find the index of the minimum value 
lowest_index = np.argmin(y)
lowest_p = p[lowest_index]
lowest_y = y[lowest_index]

# Reporting results for part (b) and part (c)
print(f"(b) The value of p that minimizes the error: {lowest_p:.2f}")
print(f"(c) The minimum error: {lowest_y:.2f}")

# Plotting the function and marking the minimum 
fig, ax = plt.subplots()

ax.plot(p, y, color='#0072B2')

# Legend showing minimum point answers 
ax.scatter(lowest_p, lowest_y, color='#D55E00', zorder=5,
            label=f'Minimum Point\n(p = {lowest_p:.2f}, error = {lowest_y:.2f})')
ax.set_title('Model Error Function $f(p)$')
ax.set_xlabel('Model Parameter (p)')
ax.set_ylabel('Error $f(p)$')

ax.legend()
ax.grid(True)


plt.show()
```
